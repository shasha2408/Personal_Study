###### 시리즈와 딕셔너리 자료형
시리즈 객체는 라벨 값에 의햇 인덱싱이 가능하므로 실질적으로 인덱스 라벨값을 키로 가지는 딕셔너리 자료형과 같다고 볼 수 있다. 따라서 딕셔너리 자료형에서 제공하는 in 연산도 가능하고 items 메서드를 사용하면 for 루프를 통해 각 원소의 키와 값을 접근할 수 있다.

###### 시리즈의 속성

s.value : 값 

###### 시리즈와 딕셔너리 자료형

"서울" in s : 인덱스 라벨 중에 서울이 있는가

for k, v in s.items() :   
    print(k, v) : key 와 value를 번갈아 가며 출력한다. 

```python
# Problem 4.1.1 
s1 = pd.Series([10, 20, 30, 40], index = ['광주', '대구', '대전', '서울'])
s2 = pd.Series([10, 30, 40, 70], index = ['제주', '원주', '광주', '서울'])

ans1 = s1 - s2
ans1[ans1.notnull()]
```

#### Chap4.3 : 데이터 고급 인덱싱
데이터프레임에서 특정한 데이터만 골라내는 것을 인덱싱이라고 한다. 앞 절에서는 라벨, 라벨 리스트, 인덱스 데이터, 슬라이스의 3가지 인덱싱 값을 사용하여 인덱싱ㅇ르 하는 방법을 공부하였다. 그런데 Pandas는 numpy행렬과 같이 쉼표를 사용한 (행 인덱스, 열 인덱스) 형식의 2차원 인덱싱을 지원하기 위해 다음과 같은 특별한 인덱서 속성도 제공한다 
- loc : 라벨 값 기반의 2차원 인덱싱 
- iloc : 순서를 나타내는 정수 기반의 2차원 인덱싱 

```python
# Problem 4.3.1
df = pd.DataFrame(np.random.randint(1,25, size = 25).reshape(5,5), columns = ['A', 'B', 'C', 'D', 'E'])

ans1 = df.iloc[1, 3]
display(ans1)

ans2 = df.iloc[:, 3]
display(ans2)

ans3 = df.iloc[1:2, 3:4]
display(ans3)

ans4 = df.loc[df.A < 15, :]
display(ans4)
```

#### Chap4.4 : 데이터프레임의 데이터 조작

##### 데이터 갯수 세기
가장 간단한 데이터 분석은 데이터의 갯수를 세는 것이다. coount 메서드를 사용한다.  
데이터프레임에서는 각 열마다 별도로 데이터 갯수를 센다. 데이터에서 값이 누락된 부분을 찾을 때 유용하다. 

```python
# Problem 4.4.1
import seaborn as sns 
titanic = sns.load_dataset('titanic')
titanic.count()
```

##### 카테고리 값 세기
시리즈의 값이 정수, 문자열, 카테고리 값인 경우에는 value_counts 메서드로 각각의 값이 나온 횟수를 셀 수 있다.  
데이터프레임에는 value_counts 메서드가 없으므로 각 열마다 별도로 적용해야 한다. 

##### 정렬 
데이터를 정렬하려면 sort_index 메서드 sort_values 메서드를 사용한다. sort_index 메서드는 인덱스 값을 기준으로, sort_values 메서드는 데이터 값을 기준으로 정렬한다. 데이터프레임에서 sort_values 메서드를 사용하려면 by 인수로 정렬 기준이 되는 열을 지정해 주어야 한다. by 인수에 리스트 값을 넣으면 이 순서대로 정렬 기준의 우선 순위가 된다. 즉, 리스트의 첫번째 열을 기준으로 정렬한 후 동일한 값이 나오면 그 다음 열로 순서를 따지게 된다. 

```python
# Problem 4.4.2
titanic.sort_values(by = ['sex', 'age', 'class', 'alive']).head()
```

##### 행/열 합계
행과 열으 ㅣ합계를 구할 때는 sum(axis) 메서드를 사용한다. axis 인수에는 합계로 인해 없어지는 방향축을 지정한다. 
- 행방향 : axis = 1
- 열방향 : axis = 0

```python
# Problem 4.4.3
ans1 = titanic['age'].mean()
ans2 = titanic.loc[titanic.sex == 'female', 'age'].mean()
ans3 = titanic.loc[(titanic.sex == 'female') & (titanic.pclass == 1), 'age'].mean()
print("%.2f, %.2f, %.2f" %(ans1, ans2, ans3))
```

##### apply 변환
행이나 열 단위로 더 복잡한 처리를 하고 싶을 때는 apply 메서드를 사용한다. 인수로 행 또는 열을 받는 함수를 apply 메서드의 인수로 넣으면 각 열을 반복하여 그 함수에 적용시킨다. 

```python
# Problem 4.4.4 
titanic['category1'] = titanic.apply(lambda x : x.sex if x.age > 20 else "child", axis = 1)
titanic.head()
```

```python
# Problem 4.4.5
titanic['age'] = titanic['age'].fillna(titanic.age.mean())
titanic.head()
```

##### 실수 값을 카테고리 값으로 변환 
실수 값을 크기 기준으로 하여 카테고리 값으로 변환하고 싶을 때는 다음과 같은 명령을 사용한다.
- cut : 실수 값의 경계선을 지정하는 경우 
- qcut : 갯수가 똑같은 구간으로 나누는 경우 
cut 명령이 반환하는 값은 Categorical 클래스 객체이다. 이 객체는 cateogires 속성으로 라벨 문자열을, codes 속성으로 정수로 인코딩한 카테고리 값을 가진다. 따라서 위 데이터프레임의 결과는 문자열이 아니고, 이를 문자열로 만드려면 astype 메서드를 사용해야 한다. 

```python
# Problem 4.4.7
bins = [1, 20, 30, 50, 70, 100]
labels = ["미성년자", "청년", "중년", "장년", "노년"]
titanic['age_cat'] = pd.cut(titanic.age, bins = bins, labels = labels)
titanic.head()
```

```python
# Problem 4.4.8
titanic['category3'] = pd.cut(titanic.age, bins = bins, labels = labels)
titanic['category3'] = titanic.apply(lambda x : "미성년자" if x.age < 20 else titanic.category3.astype('str') + x.sex, axis = 1)
titanic.head()
```
