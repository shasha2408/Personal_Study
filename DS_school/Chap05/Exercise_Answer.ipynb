```python
import numpy as np
import matplotlib as mlt
import matplotlib.pyplot as plt

X = np.linspace(-2, 2, 500)
E, C, S = np.exp(X), np.cos(X), np.sin(X)
plt.plot(X, E, label = 'exponential', ls = '--', c = 'b')
plt.plot(X, C, label = 'cosine', c = 'g')
plt.plot(X, S, label = 'sine', ls = ':', c = 'k')
plt.title("Multiple Graph")
plt.legend(loc = 2)
plt.show()
```

```python
X = np.linspace(-3, 3, 500)
y1 = np.cos(2*np.pi*X)*np.exp(-X)
y2 = np.sin(2*np.pi*X)*np.exp(-X)
y3 = (1/(2*np.pi))*np.exp(-(X**2)/2)
y4 = np.cos(X)+np.sin(X)

plt.subplot(221)
plt.plot(X, y1, ls = '--')
plt.title("Figure1")
plt.xlabel("X")
plt.ylabel("y1")

plt.subplot(222)
plt.plot(X, y2, ls = '--')
plt.title("Figure2")
plt.xlabel("X")
plt.ylabel("y2")

plt.subplot(223)
plt.plot(X, y3, ls = '--')
plt.title("Figure3")
plt.xlabel("X")
plt.ylabel("y3")

plt.subplot(224)
plt.plot(X, y4, ls = '--')
plt.title("Figure4")
plt.xlabel("X")
plt.ylabel("y4")

plt.show()
```
