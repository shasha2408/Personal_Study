# Problem 4344

문제 :  
대학생 새내기들의 90%는 자신이 반에서 평균은 넘는다고 생각한다. 당신은 그들에게 슬픈 진실을 알려줘야 한다.

입력 :  
첫째 줄에는 테스트 케이스의 개수 C가 주어진다.

둘째 줄부터 각 테스트 케이스마다 학생의 수 N(1 ≤ N ≤ 1000, N은 정수)이 첫 수로 주어지고, 이어서 N명의 점수가 주어진다. 점수는 0보다 크거나 같고, 100보다 작거나 같은 정수이다.

출력 :  
각 케이스마다 한 줄씩 평균을 넘는 학생들의 비율을 반올림하여 소수점 셋째 자리까지 출력한다.

풀이 : 
입력받은 test_case에 num_of_st를 분리하고, 평균을 계산하여 mean 변수에 저장한다.  
After doing that, use [i for i if] to make list making output condition.


```python
T = int(input())

for _ in range(T) : 
    test_case = list(map(int, input().split()))
    num_of_st = test_case[0]    #spliting first index of test_case 
    del test_case[0]
    mean = sum(test_case)/num_of_st
    num_of_ab = len([i for i in test_case if i > mean])
    print("{}%".format(format(num_of_ab/num_of_st*100, ".3f")))
```

# Problem 4673

문제 : 

셀프 넘버는 1949년 인도 수학자 D.R. Kaprekar가 이름 붙였다. 양의 정수 n에 대해서 d(n)을 n과 n의 각 자리수를 더하는 함수라고 정의하자. 예를 들어, d(75) = 75+7+5 = 87이다.

양의 정수 n이 주어졌을 때, 이 수를 시작해서 n, d(n), d(d(n)), d(d(d(n))), ...과 같은 무한 수열을 만들 수 있다. 

예를 들어, 33으로 시작한다면 다음 수는 33 + 3 + 3 = 39이고, 그 다음 수는 39 + 3 + 9 = 51, 다음 수는 51 + 5 + 1 = 57이다. 이런식으로 다음과 같은 수열을 만들 수 있다.

33, 39, 51, 57, 69, 84, 96, 111, 114, 120, 123, 129, 141, ...

n을 d(n)의 생성자라고 한다. 위의 수열에서 33은 39의 생성자이고, 39는 51의 생성자, 51은 57의 생성자이다. 생성자가 한 개보다 많은 경우도 있다. 예를 들어, 101은 생성자가 2개(91과 100) 있다. 

생성자가 없는 숫자를 셀프 넘버라고 한다. 100보다 작은 셀프 넘버는 총 13개가 있다. 1, 3, 5, 7, 9, 20, 31, 42, 53, 64, 75, 86, 97

10000보다 작거나 같은 셀프 넘버를 한 줄에 하나씩 출력하는 프로그램을 작성하시오.

출력 : 

10,000보다 작거나 같은 셀프 넘버를 한 줄에 하나씩 증가하는 순서로 출력한다.

풀이 : 

dn(n) 함수는 for loop에서 각 숫자를 읽어들여 add에 저장하고 n에 더하여 새로운 n을 업데이트 한다.   
이 과정을 for i in range(1, 10001) : 에 대입하여 dn_list에 저장하게 되면, 셀프 넘버를 제외한 list가 만들어진다.  
최종적으로 i for i 구문의 조건 수식을 이용하여 1 ~ 10000의 list에 dn_list의 원소를 제외하면,  
셀프 넘버의 list가 생성된다.     

```python
dn_list = []

def dn(n) :
    add = 0
    for i in range(len(str(n))) :
        add += int(str(n)[i])  # make dn number adding each number
    n += add
    return n

for i in range(1, 10001) :
    dn_list.append(dn(i)) # for all number make dn_list through dn()
    
for ans in [num for num in range(1, 10001) if num not in dn_list] :
    print(ans)
```

# Problem 4948

```python
def prime_list(n): 
    if n == 1 :
        return False 
    else :
        m = int(n ** 0.5)
        for i in range(2, m + 1):
            if n % i == 0 :
                return False
        return True

while True :
    N = int(input()); count = 0
    if N == 0 : 
        break
    for ans in range(N+1, 2*N + 1) :
        if prime_list(ans) :
            count += 1
    print(count)
```
