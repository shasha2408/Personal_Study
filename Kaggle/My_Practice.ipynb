```python
def least_difference(a, b, c) : 
    """Return the smallest difference between any two numbers among a, b and c.

    >>> least_difference(1, 5, -5)
    4
    """
    diff1 = abs(a - b)
    diff2 = abs(b - c)
    diff3 = abs(a - c)
    return min(diff1, diff2, diff3) 
```

```python
def mult_by_five(x) : 
    return 5 * x

def call(fn, arg) : 
    """Call fn on arg"""
    return fn(arg)

def squared_call(fn, arg) :
    """Call fn on the result of calling fn on arg"""
    return fn(fn(arg))

print(
    call(mult_by_five, 1),
    squared_call(mult_by_five, 1),
    sep = '\n'
)
```

```python
# Reading dataframe

import pandas as pd 

house_file_path = 'Datasets/house_price_prediction/train.csv'

house_data = pd.read_csv(house_file_path)

house_data.describe()
```

```python
house_data.columns
```

```python
y = house_data.SalePrice

y.head()
```

```python
house_data.isnull().sum().sort_values()
```

```python
house_features = ['GrLivArea', 'LotArea', 'GarageArea', 'TotalBsmtSF']

X = house_data[house_features]
```

```python
from sklearn.tree import DecisionTreeRegressor

house_model = DecisionTreeRegressor(random_state = 1)

# Fit model

house_model.fit(X, y)

# Make predictions

house_model.predict(X.head())
```

```python
from sklearn.metrics import mean_absolute_error

predict_y = house_model.predict(X)
mean_absolute_error(y, predict_y)
```

```python
from sklearn.model_selection import train_test_split

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)

# Define model

house_model = DecisionTreeRegressor()

# Fit model

house_model.fit(train_X, train_y)

# get predicted prices on validation data

val_predict = house_model.predict(val_X)
print(mean_absolute_error(val_y, val_predict))
```

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)

# compare MAE with differing values of max_leaf_nodes
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

```python
from sklearn.ensemble import RandomForestRegressor

forest_model = RandomForestRegressor(random_state = 1)
forest_model.fit(train_X, train_y)

pred_y = forest_model.predict(val_X)
print(mean_absolute_error(val_y, pred_y))
```
